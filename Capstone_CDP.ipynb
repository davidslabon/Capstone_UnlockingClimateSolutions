```python
# data handling
import pandas as pd
import numpy as np
import os
```

```python
def get_data(path, filename_start):
    '''a function to store the content of a directory into a pd dataframe'''
    
    # checking the contents of the directory using the os-module. 
    files = [
        file for file in os.listdir(path) 
        if file.startswith(filename_start)
        ]
    
    print(files)  
    
    # iterate through files and add to the data frame
    all_data = pd.DataFrame()
    for file in files:
        current_data = pd.read_csv(path+"/"+file)
        all_data = pd.concat([all_data, current_data])
        
    print(f'A dataframe with {all_data.shape[0]} rows and {all_data.shape[1]} columns has been created!')
    
    return all_data
```

```python
dfc = get_data(
    path='data/cdp-unlocking-climate-solutions/Cities/Cities Responses',
    filename_start='20'
    )
```

```python
def question_to_col(data, QuestionNumber, ColumnNumber):
    small_df = pd.DataFrame(columns=[QuestionNumber])
    small_df[QuestionNumber] = data['Response Answer'].loc[(data['Question Number'] == QuestionNumber) & (data['Column Number'] == int(ColumnNumber))]
    return small_df
```
